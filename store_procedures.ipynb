# task 2 ManageBooking

```python
stored_procedure_query="""create procedure ManageBooking()
begin
select cast(b.BookingSlot as char), concat(b.GuestFirstName, b.GuestLastName) as Guest_name, concat('Assogned to: ', e.name,' [', e.role,']')
FROM Bookings b left join Employees e on b.EmployeeID = e.EmployeeID
order by b.BookingSlot;
end
"""

cursor.execute(stored_procedure_query)
cursor.callproc('ManageBooking')
results2 = next(cursor.stored_results())
results2.fetchall( )
```

# task 3 UpdateBooking

```python
# cursor.execute('drop procedure UpdateBooking;')
stored_procedure_query="""create  procedure UpdateBooking(bookingid int, bookingtime varchar(8))
begin
update Bookings b set b.BookingSlot = bookingtime where b.BookingID = bookingid ;
end
"""

cursor.execute(stored_procedure_query)
args = (1,'10:00:00')
cursor.callproc('UpdateBooking',args)
cursor.execute('select bookingid,cast(BookingSlot as char), concat(GuestFirstName, GuestLastName) as Guest_name from Bookings;')
results3 = cursor.fetchall()

for result in results3:
    print(result)
```

# task 4 AddBooking

```python
stored_procedure_query="""create  procedure AddBooking(TableNo INT,
GuestFirstName VARCHAR(100) ,
GuestLastName VARCHAR(100) ,
BookingSlot TIME ,
EmployeeID INT)
begin
insert into Bookings(TableNo, GuestFirstName, 
GuestLastName, BookingSlot, EmployeeID) values (TableNo, GuestFirstName, GuestLastName, BookingSlot, EmployeeID) ;
end
"""

cursor.execute(stored_procedure_query)
args = (8, 'Dia', 'Pin', '21:00:00', 5)
cursor.callproc('AddBooking',args)
cursor.execute('select bookingid,cast(BookingSlot as char), concat(GuestFirstName, GuestLastName) as Guest_name from Bookings;')
results4 = cursor.fetchall()

for result in results4:
    print(result)
```
